```python
# json 파일 가져오기
import json

# JSON 파일에서 딕셔너리 읽기
with open('data/text_data/output_text_valid.json', 'r') as file:
    data_loaded = json.load(file)
```

```python
# 검증 데이터 분할 입력값과 라벨값

model_input = [data_loaded[f'{i}'][0] for i in range(len(data_loaded))]
labels = [data_loaded[f'{i}'][1] for i in range(len(data_loaded))]
```

```python
import torch
from transformers import T5TokenizerFast, T5ForConditionalGeneration
from transformers import GPT2LMHeadModel, PreTrainedTokenizerFast, GPT2Tokenizer

models = []
tokenizers = []

models.append(T5ForConditionalGeneration.from_pretrained('models/t5/transfer_1/model/'))
tokenizers.append(T5TokenizerFast.from_pretrained('models/t5/transfer_1/model/tokenizer/'))

models.append(T5ForConditionalGeneration.from_pretrained('models/t5/transfer_1_large/model/'))
tokenizers.append(T5TokenizerFast.from_pretrained('models/t5/transfer_1_large/model/tokenizer/'))

models.append(GPT2LMHeadModel.from_pretrained('models/gpt2/gpt2_base_1/models/'))
tokenizers.append(GPT2Tokenizer.from_pretrained('models/gpt2/gpt2_base_1/models/tokenizer/'))

models.append(GPT2LMHeadModel.from_pretrained('models/gpt2/kogpt2_1/model'))
tokenizers.append(PreTrainedTokenizerFast.from_pretrained("models/gpt2/kogpt2_1/model/tokenizer/",
                                    bos_token='</s>', eos_token='</s>', unk_token='<unk>',
                                    pad_token='<pad>', mask_token='<mask>'))



device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def gentext(text_input, model, tokenizer, device):

    # 입력 텍스트 준비
    if model.config.model_type != 't5':
        print(f'모델 타입: {model.config.model_type}')
        # GPT-2용 프롬프트
        prompt = f"입력값: {text_input}\n출력값:"
    else:
        print(f'모델 타입: {model.config.model_type}')
        # T5용 프롬프트
        prompt = text_input

    # 입력 텍스트를 토큰화
    inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True)

    # 입력 데이터와 attention mask를 GPU로 이동
    input_ids = inputs['input_ids'].to(device)
    attention_mask = inputs['attention_mask'].to(device)

    # 모델을 GPU로 이동
    model.to(device)
    # 모델과 토크나이저를 GPU로 이동
    model.to(device)
    input_ids = input_ids.to(device)

    # 모델 타입에 따라 다른 인자를 설정
    with torch.no_grad():
        if model.config.model_type == 'gpt2':
            # GPT-2의 텍스트 생성 설정
            outputs = model.generate(
                input_ids,
                attention_mask=attention_mask,  # Attention mask 추가
                max_new_tokens=64,
                num_beams=1,
                no_repeat_ngram_size=30,
                top_k=50,
                top_p=0.95,
                temperature=1,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id  # pad_token_id 설정
            )
        elif model.config.model_type == 't5':
            # T5의 텍스트 생성 설정
            outputs = model.generate(
                input_ids,
                attention_mask=attention_mask,  # Attention mask 추가
                max_new_tokens=64,
                num_beams=1,
                pad_token_id=tokenizer.eos_token_id  # pad_token_id 설정
            )
    # 예측 결과 디코딩
    predicted_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    try:
        out = predicted_text.split('출력값:')[1]
    except:
        out = predicted_text

    return out
```

```python
results_t5_base = []
results_t5_large = []
results_gpt2 = []
results_kogpt2 = []

for i,l in enumerate([results_t5_base, results_t5_large, results_gpt2, results_kogpt2]):
    for j in model_input:
        l.append(gentext(j, models[i], tokenizers[i], device))
```

```python
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score
from rouge_score import rouge_scorer
from loguru import logger

# 로그로 저장 (파일 로깅 설정) 정보만 출력
# info 출력 안함
logger.add('logs/LMmodel_BLEU.log', format="{message} - {time}", level="INFO", backtrace=False, diagnose=False)

n_gram_weights = [
    (1, 0, 0, 0),  # BLEU-1
    (0.5, 0.5, 0, 0),  # BLEU-2
    (0.33, 0.33, 0.33, 0),  # BLEU-3
    (0.25, 0.25, 0.25, 0.25)  # BLEU-4
]

# BLEU 점수 계산
for n, w in enumerate(n_gram_weights):
    # 결과 로깅
    logger.info(f"BLEU-{n+1} score for T5 Base: {corpus_bleu([[i] for i in labels], results_t5_base, weights=w, smoothing_function=SmoothingFunction().method0):.4f}")
    logger.info(f"BLEU-{n+1} score for T5 Large: {corpus_bleu([[i] for i in labels], results_t5_large, weights=w, smoothing_function=SmoothingFunction().method0):.4f}")
    logger.info(f"BLEU-{n+1} score for KoGPT2: {corpus_bleu([[i] for i in labels], results_kogpt2, weights=w, smoothing_function=SmoothingFunction().method0):.4f}")
    logger.info(f"BLEU-{n+1} score for GPT2: {corpus_bleu([[i] for i in labels], results_gpt2, weights=w, smoothing_function=SmoothingFunction().method0):.4f}")
logger.remove()
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.translate.meteor_score import meteor_score
from loguru import logger

# 로그로 저장 (파일 로깅 설정) - 시간과 메시지만 출력
logger.add('logs/LMmodel_METEOR.log', format="{time} - {message}", level="INFO", backtrace=False, diagnose=False)

# 정규 표현식을 사용한 토큰화
tokenizer = RegexpTokenizer(r'\w+')

def tokenize_sentence(sentence):
    return tokenizer.tokenize(sentence)

for i, result in enumerate(zip(results_t5_base, results_t5_large, results_kogpt2, results_gpt2), start=1):
    if i-1 < len(labels):  # 인덱스 범위 체크
        meteor_scores = {
            "T5 Base": meteor_score([tokenizers[0](labels[i-1])], tokenizers[0](result[0])),
            "T5 Large": meteor_score([tokenizers[1](labels[i-1])], tokenizers[1](result[1])),
            "KoGPT2": meteor_score([tokenizers[2](labels[i-1])], tokenizers[2](result[2])),
            "GPT2": meteor_score([tokenizers[3](labels[i-1])], tokenizers[3](result[3]))
        }
    else:
        logger.error(f"Index {i-1} out of range for labels.")
    

# 전체 모델에 대한 METEOR 평균 점수 계산
def average_meteor_score(results, labels):
    scores = [meteor_score([tokenize_sentence(label)], tokenize_sentence(result)) for result, label in zip(results, labels)]
    return sum(scores) / len(scores)

# 각 모델에 대한 평균 METEOR 점수
avg_meteor_scores = {
    "T5 Base": average_meteor_score(results_t5_base, labels),
    "T5 Large": average_meteor_score(results_t5_large, labels),
    "KoGPT2": average_meteor_score(results_kogpt2, labels),
    "GPT2": average_meteor_score(results_gpt2, labels)
}

# 결과 로깅
for model, score in avg_meteor_scores.items():
    logger.info(f"Average METEOR score for {model}: {score:.4f}")

logger.remove()
```
